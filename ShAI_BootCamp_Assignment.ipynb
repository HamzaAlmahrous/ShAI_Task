```python
import pandas as pd
import numpy as np

# Load your dataset
# df = pd.read_csv('/content/Salaries.csv')
df = pd.read_csv('Salaries.csv')
df.head()
```

1. **Basic Data Exploration**

Number of rows and columns:

```python
print(f"Number of rows: {df.shape[0]}")
print(f"Number of columns: {df.shape[1]}")
```

Data types of each column:

```python
print("Data types of each column:")
print(df.dtypes)
```

The number of missing values in each column:

```python
print("Missing values in each column:")
print(df.isnull().sum())
```

2. **Descriptive Statistics**

```python
salary_statistics = df['TotalPay'].describe()
salary_mode = df['TotalPay'].mode()
salary_std_dev = df['TotalPay'].std()
salary_median = df['TotalPay'].median()
salary_range = df['TotalPay'].max() - df['TotalPay'].min()

print("\nStatistics Description for Salary:")
print(salary_statistics)
print("\nMode of Salary:")
print(salary_mode)
print("\nMedian of Salary:")
print(salary_median)
print("\nStandard Deviation of Salary:")
print(salary_std_dev)
print("\nRange of Salaries:")
print(salary_range)
```

3. **Data Cleaning**

we can remove all the rows that has a missing values or we can fill the missing values wit the mean

```python
df_cleaned = df.dropna()
```

```python
df_cleaned_mean = df.copy()

print("\nMissing values in each column after imputation:")
print(df_cleaned_mean.isnull().sum())
```

```python
df_cleaned_mean['BasePay'].fillna(df_cleaned_mean['BasePay'].mean, inplace=True)
df_cleaned_mean['OvertimePay'].fillna(df_cleaned_mean['OvertimePay'].mean, inplace=True)
df_cleaned_mean['OtherPay'].fillna(df_cleaned_mean['OtherPay'].mean, inplace=True)
df_cleaned_mean['Benefits'].fillna(df_cleaned_mean['Benefits'].mean, inplace=True)
df_cleaned_mean['Notes'].fillna(df_cleaned_mean['Notes'].mean, inplace=True)
df_cleaned_mean['Status'].fillna(df_cleaned_mean['Status'].mean, inplace=True)
```

```python
print("\nMissing values in each column after imputation:")
print(df_cleaned_mean.isnull().sum())
```

4. **Basic Data Visualization**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_cleaned_mean['TotalPay'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Salaries')
plt.xlabel('Total Pay (Salary)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
department_counts = df_cleaned_mean['JobTitle'].value_counts().head(20)
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Proportion of Employees in Different Departments')
plt.show()
```

5. **Grouped Analysis**

```python
average_salary_by_jobtitle = df_cleaned_mean.groupby('JobTitle')['TotalPay'].mean().sort_values(ascending=False)

print("Average Salary by Job Title:")
print(average_salary_by_jobtitle)
```

6. **Simple Correlation Analysis**

```python
correlation = df['TotalPay'].corr(df['OtherPay'])
print(f"Correlation between {'TotalPay'} and {'OtherPay'}: {correlation}")
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['OtherPay'], y=df['TotalPay'])
plt.title(f'Scatter Plot: TotalPay vs. OtherPay')
plt.xlabel('OtherPay')
plt.ylabel('TotalPay')
plt.show()
```
